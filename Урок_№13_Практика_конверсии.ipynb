```python
# сделаем свою функцию - симуляция монетки
def coin(p): # int = integer - целое число возвращает 0 или 1
  return int(rnd.random() < p) # если случайное число меньше заданного p, возвращай 1. Иначе 0
# если
```

```python
# сделаем свою функцию для нескольких монет
def bag_coins(n, p): # сколько выпало орлов на n монетах, если каждая имеет вероятность p
  return sum( [coin(p) for _ in range(n)] ) # повторяем n раз. Подсчитываем сумму единичек
```

```python
# Проводим настоящий эксперимент
w = 20 # размер тестовой группы. 
real = rnd.choice([0.5, 0.8]) # Существует РЕАЛЬНАЯ конверсия. Но мы её не узнаем. В переменную real поместили либо 0.5, либо 0.8
# настоящий тест
res = bag_coins(w, real) # Звоним w людям. Смотрим, какая среди них получилась конверсия?
```

```python
rnd.seed(7) # строчка обеспечивает повторяемость случайности от запуска к запуску 

w = 20
p1 = 0.8

H0 = [bag_coins(w, 0.5) for _ in range(10000)] # все монетки честные
H1 = [bag_coins(w, p1) for _ in range(10000)] # все монетки нечестные
plt.hist(H0, bins=14, color='green', label='H0')
plt.hist(H1, bins=14, color='red', label='H1')

real = rnd.choice([0.5, p1]) # простой вариант, когда альтернативная гипотеза известна, мы можем построить красную гистограмму
# настоящий тест
res = bag_coins(w, real)

plt.plot([res, res], [0, 2600], color='black', label ='Result AB-test')

plt.legend();
```

```python
w = 200
p1 = rnd.choice([0.52, 0.57, 0.58, 0.62, 0.68, 0.7, 0.8]) # следующий уровень, ближе к реальности - альтернаяивная гипотеза не определена. У нас есть только зелёная гистограмма 

real = rnd.choice([0.5, p1])
# настоящий тест
res = bag_coins(w, real)
res

H0 = [bag_coins(w, 0.5) for _ in range(10000)] # все монетки честные


H0_left  = [e for e in H0 if e < res] # выделяем все симуляции, которые меньше res
H0_right = [e for e in H0 if e >= res] # выделяем все симуляции, которые больше или равны res

plt.hist(H0_left, bins=14, color='green', label='H0')
plt.hist(H0_right, bins=14, color='green', label='H0')

print("P_value: ", len(H0_right) /  len(H0))

plt.plot([res, res], [0, 1500], color='black', label ='Result AB-test')

plt.legend();
```

# Основной вопрос: может ли такой результат получиться случайно при верной Н0?

# Основной вопрос: появилась ли чёрная линия из зелёной гистограмме?

Что если наша Н0 верна, тогда с какой вероятностью, мы бы получили такое отклонение или больше? - ответ: p_value

```python
p_value = len(H0_right) /  len(H0)  # доля гистограммы справа от чёрной линии
if p_value  < 0.05: # cтандарт
  print('Это, скорее всего, Н1')
else:
  print('Это, скорее всего, H0')
```

P_value - вероятность получить такое же отклонение или больше, если верна нулевая гипотеза.
